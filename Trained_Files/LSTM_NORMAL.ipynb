```python
# -*- coding: utf-8 -*-
"""LSTM_implementation.ipynb"""

!pip install numpy pandas scikit-learn tensorflow joblib matplotlib

import pandas as pd
import numpy as np
import tensorflow as tf
import random

# full precision display
pd.set_option('display.float_format', lambda x: f"{x:.16E}")

# reproducibility
seed = 42
random.seed(seed)
np.random.seed(seed)
tf.random.set_seed(seed)

df = pd.read_csv(
    "/content/C12_final_with_tseconds (1) (2).csv",
    parse_dates=["timestamp"],
    dtype={"bias_float64": "float64", "drift_float64": "float64"}
)

df = df.sort_values("timestamp").reset_index(drop=True)
print(df.info())
print(df.head())

print(df.columns.tolist())

# interpolate drift
df["drift_float64"] = df["drift_float64"].interpolate("linear", limit_direction="both")

# reindex to 30-second grid
start, end = df["timestamp"].iloc[0], df["timestamp"].iloc[-1]
idx = pd.date_range(start, end, freq="30S")

df = df.set_index("timestamp").reindex(idx)
df.index.name = "timestamp"
df = df.reset_index()

# interpolate drift only (bias untouched)
df["drift_float64"] = df["drift_float64"].interpolate("linear", limit_direction="both")

# ensure bias no NaN
if df["bias_float64"].isna().sum() > 0:
    print("WARNING: Missing bias detected — filling with nearest value.")
    df["bias_float64"] = df["bias_float64"].ffill().bfill()

# train/test split
test_start = pd.to_datetime("2024-01-20 00:00:00")

train_df = df[df["timestamp"] < test_start].copy()
test_df  = df[df["timestamp"] >= test_start].copy()

# validation from end of train
val_size = int(len(train_df) * 0.1)
val_df = train_df.iloc[-val_size:].copy()
train_df = train_df.iloc[:-val_size].copy()

from sklearn.preprocessing import StandardScaler

scaler_bias  = StandardScaler()
scaler_drift = StandardScaler()

scaler_bias.fit( train_df[["bias_float64"]].values )
scaler_drift.fit(train_df[["drift_float64"]].values)

def transform(df):
    Xb = scaler_bias.transform(df[["bias_float64"]].values).astype("float32")
    Xd = scaler_drift.transform(df[["drift_float64"]].values).astype("float32")
    return np.hstack([Xb, Xd])

X_train_all = transform(train_df)
X_val_all   = transform(val_df)
X_test_all  = transform(test_df)

def make_windows(X, window=60):
    Xs, Ys = [], []
    for i in range(len(X) - window):
        Xs.append(X[i:i+window])
        Ys.append(X[i+window][0])  # target = scaled bias
    return np.array(Xs), np.array(Ys)

window_size = 60
Xtr, ytr = make_windows(X_train_all, window_size)
Xval, yval = make_windows(X_val_all, window_size)
Xte, yte = make_windows(X_test_all, window_size)

print(Xtr.shape, ytr.shape)

from tensorflow.keras import layers, models

def build_lstm(window, features):
    inp = layers.Input(shape=(window, features))
    x = layers.LSTM(128, return_sequences=True)(inp)
    x = layers.LSTM(64)(x)
    x = layers.Dropout(0.2)(x)
    x = layers.Dense(32, activation='relu')(x)
    out = layers.Dense(1)(x)
    return models.Model(inp, out)

model = build_lstm(window_size, Xtr.shape[2])
model.compile(optimizer="adam", loss="mse", metrics=["mae"])
model.summary()

from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint

es = EarlyStopping(patience=10, restore_best_weights=True)
mc = ModelCheckpoint("lstm_best.h5", save_best_only=True)

history = model.fit(
    Xtr, ytr,
    validation_data=(Xval, yval),
    epochs=100,
    batch_size=64,
    callbacks=[es, mc],
    verbose=2
)

y_pred_scaled = model.predict(Xte).flatten()

# invert scaling
y_pred = scaler_bias.inverse_transform(y_pred_scaled.reshape(-1,1)).flatten()
y_true = scaler_bias.inverse_transform(yte.reshape(-1,1)).flatten()

out = pd.DataFrame({
    "timestamp": test_df["timestamp"].iloc[window_size: window_size+len(y_pred)].values,
    "true_bias": y_true,
    "pred_bias": y_pred
})

out.to_csv("C12_LSTM_predictions.csv", index=False, float_format="%.16E")

from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

mae = mean_absolute_error(y_true, y_pred)
rmse = np.sqrt(mean_squared_error(y_true, y_pred))
r2 = r2_score(y_true, y_pred)

print("MAE (s):", mae)
print("RMSE (s):", rmse)
print("R2 Score:", r2)

# PLOTS
import matplotlib.pyplot as plt

plt.figure(figsize=(15,4))
plt.plot(df["timestamp"], df["bias_float64"], linewidth=0.8)
plt.title("Clock Bias (float64)")
plt.xlabel("Time")
plt.ylabel("Bias (s)")
plt.grid(True)
plt.show()

plt.figure(figsize=(15,4))
plt.plot(df["timestamp"], df["drift_float64"], linewidth=0.8, color='orange')
plt.title("Clock Drift (float64)")
plt.xlabel("Time")
plt.ylabel("Drift")
plt.grid(True)
plt.show()

plt.figure(figsize=(10,4))
plt.hist(df["bias_float64"], bins=100, alpha=0.8)
plt.title("Bias Distribution")
plt.grid(True)
plt.show()

plt.figure(figsize=(10,4))
plt.hist(df["drift_float64"], bins=100, color="orange", alpha=0.8)
plt.title("Drift Distribution")
plt.grid(True)
plt.show()

plt.figure(figsize=(15,4))
plt.plot(train_df["timestamp"], train_df["bias_float64"], label="Train", linewidth=0.7)
plt.plot(test_df["timestamp"], test_df["bias_float64"], label="Test", linewidth=0.7)
plt.title("Train vs Test")
plt.legend()
plt.grid(True)
plt.show()

plt.figure(figsize=(15,4))
plt.plot(out["timestamp"], out["true_bias"], label="True", linewidth=1)
plt.plot(out["timestamp"], out["pred_bias"], label="Predicted", linewidth=1)
plt.legend()
plt.grid(True)
plt.show()

out["error"] = out["pred_bias"] - out["true_bias"]

plt.figure(figsize=(15,4))
plt.plot(out["timestamp"], out["error"], color='red')
plt.title("Prediction Error")
plt.grid(True)
plt.show()

plt.figure(figsize=(8,4))
plt.hist(out["error"], bins=100, color='crimson', alpha=0.8)
plt.title("Error Distribution")
plt.grid(True)
plt.show()

rolling_mae = np.abs(out["error"]).rolling(200).mean()

plt.figure(figsize=(12,4))
plt.plot(out["timestamp"], rolling_mae)
plt.title("Rolling MAE (200 window)")
plt.grid(True)
plt.show()

plt.figure(figsize=(6,6))
plt.scatter(out["true_bias"], out["pred_bias"], s=3, alpha=0.5)
plt.title("True vs Predicted")
plt.grid(True)
plt.show()
```

```python
# Single-cell corrected GRU pipeline (end-to-end, safe for high-precision clock data)
# - Uses columns: timestamp, bias_float64, drift_float64
# - Keeps float64 precision throughout (TensorFlow floatx set to float64)
# - Safe interpolation, reindexing to 30s, train/val/test chronological split
# - Proper windowing that returns aligned timestamps for predictions
# - Saves predictions with full precision
#
# Requirements: pandas, numpy, scikit-learn, tensorflow, scipy
# Run this in one notebook cell.

import os, random, math
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from scipy.stats import pearsonr
import tensorflow as tf
from tensorflow.keras import layers, models, callbacks, optimizers

# -----------------------------
# Repro / environment
# -----------------------------
seed = 42
random.seed(seed)
np.random.seed(seed)
tf.random.set_seed(seed)
# Use float64 everywhere to preserve precision
tf.keras.backend.set_floatx('float64')
pd.set_option('display.float_format', lambda x: f"{x:.16E}")

# -----------------------------
# Parameters (edit paths/dates if needed)
# -----------------------------
INPUT_CSV = "/content/C12_final_with_tseconds (1) (2).csv"   # your CSV file name
TEST_START = pd.to_datetime("2024-01-20 00:00:00")  # start of test day
WINDOW_SIZE = 60    # number of timesteps per input window (60*30s = 30 minutes)
VAL_RATIO = 0.10    # fraction of train to use as validation
BATCH_SIZE = 64
EPOCHS = 100
LR = 1e-3
OUT_CSV = "out.csv"

# -----------------------------
# Load CSV and basic checks
# -----------------------------
df = pd.read_csv(INPUT_CSV, parse_dates=["timestamp"])
required = {"bias_float64", "drift_float64", "timestamp"}
if not required.issubset(df.columns):
    raise RuntimeError(f"Input CSV must contain columns: {required}. Found: {df.columns.tolist()}")

df = df.sort_values("timestamp").reset_index(drop=True)
print("Loaded rows:", len(df))
print(df[["timestamp","bias_float64","drift_float64"]].iloc[:3])

# -----------------------------
# Interpolate drift only, preserve bias values (bias sensitive)
# -----------------------------
df["drift_float64"] = df["drift_float64"].interpolate(method="linear", limit_direction="both")

# -----------------------------
# Reindex to regular 30-second grid (ensures consistent windows)
# - We reindex and then INTERPOLATE drift and bias carefully.
# - For bias we only interpolate if gaps exist (warn) and then forward/backfill.
# -----------------------------
start, end = df["timestamp"].iloc[0], df["timestamp"].iloc[-1]
idx = pd.date_range(start, end, freq="30S")
df = df.set_index("timestamp").reindex(idx).rename_axis("timestamp").reset_index()

# Keep satellite column if present (fill with constant if missing)
if "satellite" in df.columns:
    df["satellite"] = df["satellite"].fillna(method="ffill").fillna(method="bfill")
else:
    df["satellite"] = "C12"

# Interpolate drift on the reindexed grid
df["drift_float64"] = df["drift_float64"].interpolate(method="linear", limit_direction="both")

# For bias: if NaNs appear due to missing epochs, fill with nearest neighbors (ffill/bfill).
# This is preferable to heavy interpolation for extremely sensitive bias; warn if any were missing.
n_missing_bias = int(df["bias_float64"].isna().sum())
if n_missing_bias:
    print(f"WARNING: {n_missing_bias} missing bias entries found after reindexing. Filling with forward/backfill.")
    df["bias_float64"] = df["bias_float64"].ffill().bfill()

# Final check
if df["bias_float64"].isna().any():
    raise RuntimeError("bias_float64 still contains NaN after fill - aborting")

# -----------------------------
# Split: 6 days for training (chronological), 7th day for testing
# -----------------------------
train_df = df[df["timestamp"] < TEST_START].copy()
test_df  = df[df["timestamp"] >= TEST_START].copy()
if len(test_df) == 0:
    raise RuntimeError("Test split empty. Check TEST_START date and dataset range.")

# Validation slice from tail of train
val_count = int(len(train_df) * VAL_RATIO)
if val_count == 0:
    val_count = max(1, int(0.05 * len(train_df)))  # ensure at least one sample
val_df = train_df.iloc[-val_count:].copy()
train_df = train_df.iloc[:-val_count].copy()

print("Train rows:", len(train_df), "Val rows:", len(val_df), "Test rows:", len(test_df))

# -----------------------------
# Scaling (fit on train only) - use StandardScaler on float64
# -----------------------------
scaler_bias = StandardScaler()
scaler_drift = StandardScaler()

scaler_bias.fit(train_df[["bias_float64"]].values.astype("float64"))
scaler_drift.fit(train_df[["drift_float64"]].values.astype("float64"))

def transform_df_to_array(df_local):
    b = scaler_bias.transform(df_local[["bias_float64"]].values.astype("float64"))
    d = scaler_drift.transform(df_local[["drift_float64"]].values.astype("float64"))
    # keep dtype float64 for TensorFlow float64 execution
    X = np.hstack([b, d]).astype("float64")
    return X

X_train_all = transform_df_to_array(train_df)
X_val_all   = transform_df_to_array(val_df)
X_test_all  = transform_df_to_array(test_df)

# -----------------------------
# Window maker (returns windows, targets, and aligned timestamps)
# target is the next-step bias (univariate)
# -----------------------------
def make_windows(Xarr, timestamps, window):
    Xw, Yw, Tw = [], [], []
    N = Xarr.shape[0]
    for i in range(N - window):
        Xw.append(Xarr[i : i + window])
        Yw.append(Xarr[i + window, 0])  # bias is column 0 (scaled)
        Tw.append(timestamps[i + window])  # timestamp aligned to target epoch
    return np.asarray(Xw, dtype="float64"), np.asarray(Yw, dtype="float64"), np.array(Tw)

Xtr, ytr, ttr = make_windows(X_train_all, train_df["timestamp"].values, WINDOW_SIZE)
Xval, yval, tval = make_windows(X_val_all, val_df["timestamp"].values, WINDOW_SIZE)
Xte, yte, tte = make_windows(X_test_all, test_df["timestamp"].values, WINDOW_SIZE)

print("Window shapes -> Xtr:", Xtr.shape, " ytr:", ytr.shape, " Xte:", Xte.shape)

# -----------------------------
# Build GRU model (float64 throughout)
# -----------------------------
n_features = Xtr.shape[2]

def build_gru(window, features):
    inp = layers.Input(shape=(window, features), dtype=tf.float64)
    x = layers.GRU(128, return_sequences=True, dtype=tf.float64)(inp)
    x = layers.GRU(64, dtype=tf.float64)(x)
    x = layers.Dropout(0.2)(x)
    x = layers.Dense(32, activation="relu", dtype=tf.float64)(x)
    out = layers.Dense(1, activation="linear", dtype=tf.float64)(x)
    model = models.Model(inp, out)
    return model

model = build_gru(WINDOW_SIZE, n_features)
model.compile(optimizer=optimizers.Adam(learning_rate=LR), loss="mse", metrics=["mae"])
model.summary()

# -----------------------------
# Train with early stopping + checkpoint
# -----------------------------
ckpt = "best_gru_model.h5"
es = callbacks.EarlyStopping(monitor="val_loss", patience=10, restore_best_weights=True)
mc = callbacks.ModelCheckpoint(ckpt, monitor="val_loss", save_best_only=True, save_weights_only=False)

history = model.fit(
    Xtr, ytr,
    validation_data=(Xval, yval),
    epochs=EPOCHS,
    batch_size=BATCH_SIZE,
    callbacks=[es, mc],
    verbose=2
)

# -----------------------------
# Predict on test (scaled space), inverse-transform to seconds (float64)
# -----------------------------
y_pred_scaled = model.predict(Xte).flatten()
y_true_scaled = yte.flatten()

y_pred_seconds = scaler_bias.inverse_transform(y_pred_scaled.reshape(-1,1)).ravel().astype("float64")
y_true_seconds = scaler_bias.inverse_transform(y_true_scaled.reshape(-1,1)).ravel().astype("float64")

# -----------------------------
# Metrics (full-precision)
# -----------------------------
mae = mean_absolute_error(y_true_seconds, y_pred_seconds)
mse = mean_squared_error(y_true_seconds, y_pred_seconds)
rmse = math.sqrt(mse)
r2 = r2_score(y_true_seconds, y_pred_seconds)
r, _ = pearsonr(y_true_seconds, y_pred_seconds)

print("\n=== EVALUATION (seconds, full precision) ===")
print("MAE     :", f"{mae:.16E}")
print("MSE     :", f"{mse:.16E}")
print("RMSE    :", f"{rmse:.16E}")
print("R^2     :", f"{r2:.12f}")
print("Pearson :", f"{r:.12f}")
print("Test sample count:", len(y_true_seconds))

# -----------------------------
# Save aligned predictions with timestamps (full precision)
# -----------------------------
out_df = pd.DataFrame({
    "timestamp": tte,
    "true_bias_seconds": y_true_seconds,
    "pred_bias_seconds": y_pred_seconds,
    "error_seconds": (y_pred_seconds - y_true_seconds)
})
out_df.to_csv(OUT_CSV, index=False, float_format="%.16E")
print("Saved predictions to:", OUT_CSV)

# -----------------------------
# Basic plots (optional, comment out if running headless)
# -----------------------------
try:
    import matplotlib.pyplot as plt
    plt.figure(figsize=(12,4))
    plt.plot(out_df["timestamp"], out_df["true_bias_seconds"], label="true", linewidth=1)
    plt.plot(out_df["timestamp"], out_df["pred_bias_seconds"], label="pred", linewidth=1)
    plt.legend()
    plt.title("Predicted vs True (seconds)")
    plt.show()

    plt.figure(figsize=(8,4))
    plt.hist(out_df["error_seconds"], bins=200)
    plt.title("Error distribution (seconds)")
    plt.show()
except Exception:
    pass

# End of cell
```
